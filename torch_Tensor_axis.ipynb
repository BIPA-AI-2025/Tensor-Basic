# Tensor shape, axis

## numpy.ndarray

```python
import numpy as np
```

### Scalar (Number)

```python
x = np.array(5)
```

```python
print(x)
print(type(x))
print('rank : ', x.ndim)
print('shape : ', x.shape)
```

### Vector

```python
x = np.array([1,2,3,5])
```

```python
print(x)
print(type(x))
print('rank : ', x.ndim)
print('shape : ', x.shape)
```

### Matrix

```python
x = np.array([
               [1,2,3,5],
               [4,5,6,7]
             ])
```

```python
print(x)
print(type(x))
print('rank : ', x.ndim)
print('shape : ', x.shape)
```

### Tensor

```python
x = np.array([  # (3, 2, 4)
               [
                 [1,2,3,5],
                 [4,5,6,9]
               ],
               [
                 [7,8,9,1],
                 [1,7,2,6]
               ],
               [
                 [7,8,9,1],
                 [1,7,2,6]
               ]
            ])
```

```python
print(x)
print(type(x))
print('rank : ', x.ndim)
print('shape : ', x.shape)
```

### numpy to torch

```python
import numpy as np
import torch

# numpy array creation
numpy_array = np.array([[1, 2, 3], [4, 5, 6]])
print(numpy_array)
print(type(numpy_array))

# numpy to torch tesnor
torch_tensor = torch.from_numpy(numpy_array)

print(torch_tensor)
print(type(torch_tensor))
```

### torch to numpy

```python
import torch

# torch tensor
torch_tensor = torch.tensor([[1, 2, 3], [4, 5, 6]])
print(torch_tensor)
print(type(torch_tensor))
print('rank : ', torch_tensor.ndim)
print('shape : ', torch_tensor.shape)

# torch tensor to numpy array
numpy_array = torch_tensor.numpy()

print(numpy_array)
print(type(numpy_array))
print('rank : ', numpy_array.ndim)
print('shape : ', numpy_array.shape)
```
